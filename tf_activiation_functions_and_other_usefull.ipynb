## Tensorflow various activation function
* plot of few activation function from Tensorflow

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Activation Function (base Tensorflow)

```python
x = np.linspace(-10,10,100)
```

```python
x_tf = tf.constant(x, dtype=tf.float32)
sess = tf.Session()
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,sess.run(tf.nn.relu(x_tf)))
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,sess.run(tf.sigmoid(x_tf)))
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,sess.run(tf.tanh(x_tf)))
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,sess.run(tf.nn.dropout(x_tf, keep_prob=0.8)))
```
